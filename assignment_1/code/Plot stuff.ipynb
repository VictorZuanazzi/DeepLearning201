```python
import matplotlib as plt
import seaborn as sns
import numpy as np
import torch
import pandas as pd
```

```python
def plot_n_save(df, title='', xlabel = 'Evaluations', ylabel='Accuracy'):
    sns.set(rc={'figure.figsize':(8,5)})
    ax = sns.lineplot(data = df)
    ax.set_title(title)
    ax.set_ylabel(ylabel)
    ax.set_xlabel(xlabel)
    fig = ax.get_figure()
    title = title.split()
    title = "_".join(title)
    fig.savefig(title + ".png")
    
```

```python
##numpy

np_path = "./np results/original parameters/"
df_acc = pd.DataFrame(data = None, columns = ["train", "test"])
df_acc["train"] = np.load(np_path+"np_accuracy_train.npy")
df_acc["test"] = np.load(np_path+"np_accuracy_test.npy")

df_loss = pd.DataFrame(data = None, columns = ["train", "test"])
df_loss["train"] = np.load(np_path+"np_loss_train.npy")
df_loss["test"] = np.load(np_path+"np_loss_test.npy")

#numpy plot and save accuracies
plot_n_save(df_acc, title = 'Numpy Accuracies')
```

```python
#numpy plot and save Loss
plot_n_save(df_loss, title = 'Numpy Loss', ylabel='Loss')
```

```python
##torch

torch_path = "./torch results/original setting/"

#load metrics
df_acc = pd.DataFrame(data = None, columns = ["train", "test"])
df_acc["train"] = np.load(torch_path+"torch_accuracy_train.npy")
df_acc["test"] = np.load(torch_path+"torch_accuracy_test.npy")

#numpy plot and save accuracies
plot_n_save(df_acc, title = 'Torch Accuracies')
```

```python
df_loss = pd.DataFrame(data = None, columns = ["train", "test"])
df_loss["train"] = np.load(torch_path+"torch_loss_train.npy")
df_loss["train"] = df_loss["train"].apply(lambda x: x.item())
df_loss["test"] = np.load(torch_path+"torch_loss_test.npy")
df_loss["test"] = df_loss["test"].apply(lambda x: x.item())

#numpy plot and save Loss
plot_n_save(df_loss, title = 'Torch Loss', ylabel='Loss')
```
